```python
#image transformation
#shifting image
pic = cv2.imread('janelle.png')
cols = pic.shape[1]
rows = pic.shape[0]
N = np.float32([[1, 0, -150], [0,1,-70]])
shifted = cv2.warpAffine(pic, N, (cols, rows))
cv2.imshow('shifted',shifted)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#image rotation
pic = cv2.imread('janelle.png')
cols = pic.shape[1]
rows = pic.shape[0]
center = (cols/2, rows/2)
angle = 45
N = cv2.getRotationMatrix2D(center, angle, 1)
rotate = cv2.warpAffine(pic, N, (cols, rows))
cv2.imshow('rotated',rotate)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#image thresholding
pic = cv2.imread('janelle.png', 0)
threshold_value = 100
(T_value, binary_threshold) = cv2.threshold(pic, threshold_value, 255, cv2.THRESH_BINARY)
cv2.imshow('binary',binary_threshold)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Guassian blur
pic = cv2.imread('janelle.png')
matrix = (7,7)
blur = cv2.GaussianBlur(pic, matrix, 0)

cv2.imshow('blurred',blur)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Median blur
pic = cv2.imread('janelle.png')
kernel = 3
median = cv2.medianBlur(pic, kernel)
cv2.imshow('median',median)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Bilateral filtering
pic = cv2.imread('janelle.png')
dimpixel = 7
color = 100
space = 100

filter = cv2.bilateralFilter(pic,dimpixel,color,space)
cv2.imshow('filter',filter)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#canny edge detector
pic = cv2.imread('janelle.png')
thresholdval1 = 50
thresholdval2 = 100

canny = cv2.Canny(pic,thresholdval1, thresholdval2)
cv2.imshow('canny',canny)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Load a video
cap = cv2.VideoCapture('video.mp4')
fourcc = cv2.VideoWriter_fourcc(*'XVID')
fps = 30
framesize = (720,480)
out = cv2.VideoWriter('sample.avi',fourcc,fps,framesize)
while(cap.isOpened()):
    ret, frame = cap.read()
    cv2.imshow('vid', frame)
    if(cv2.waitKey(1) & 0xFF == ord('q')):
        break
out.release()
cap.release()
cv2.destroyAllWindows()
```

```python
#Realtime face detection with webcam

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

videocapture = cv2.VideoCapture(0)
scale_factor = 1.3

while 1:
    ret, pic = videocapture.read()
    faces = face_cascade.detectMultiScale(pic, scale_factor, 5)
    for(x,y,w,h) in faces:
        cv2.rectangle(pic, (x, y), (x + w, y + h), (255, 0, 0), 2)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(pic, 'Me', (x, y), font, 2, (255, 255, 255), 2, cv2.LINE_AA)
    
    print("Number of faces found {} ".format(len(faces)))
    cv2.imshow('face', pic)
    k = cv2.waitKey(30) & 0xff
    if k == 2:
        break
cv2.destroyAllWindows()
```
